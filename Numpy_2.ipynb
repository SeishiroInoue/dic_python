```python
# 問題1

import numpy as np

np.random.seed(0)

mean = (-3, 0)
cov = ([1.0, 0.8], [0.8, 1.0])
dummy_ndarray = np.random.multivariate_normal(mean, cov, 500)
dummy_ndarray.shape
```

```python
# 問題2

import matplotlib.pyplot as plt

x1, x2 = np.random.multivariate_normal(mean, cov, 500).T
fig, ax = plt.subplots()
ax.set_title('scatter')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.scatter(x1, x2)
plt.show()
```

```python
# 問題3

fig, axes = plt.subplots(2)

axes[0].set_title('histogram  x1')
axes[0].set_xlabel('x1')
axes[0].set_ylabel('frequency')
axes[0].set_xlim([-5.5, 2.5])
axes[0].hist(x1, bins=50)

axes[1].set_title('histogram  x2')
axes[1].set_xlabel('x2')
axes[1].set_ylabel('frequency')
axes[1].set_xlim([-5.5, 2.5])
axes[1].hist(x1, bins=50)

plt.show()
```

```python
# 問題4

np.random.seed(0)

mean1 = (-3, 0)
mean2 = (0, -3)
cov = ([1.0, 0.8], [0.8, 1.0])

x1, y1 = np.random.multivariate_normal(mean1, cov, 500).T
x2, y2 = np.random.multivariate_normal(mean2, cov, 500).T

fig, ax = plt.subplots()
ax.set_title('scatter')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.scatter(x1, y1, label='0')
ax.scatter(x2, y2, label='1')
ax.legend()
plt.show()
```

```python
# 問題5

join_ndarray1 = np.random.multivariate_normal(mean1, cov, 500)
join_ndarray2 = np.random.multivariate_normal(mean2, cov, 500)
joins = np.concatenate((join_ndarray1, join_ndarray2))
joins.shape
```

```python
# 問題6

one = np.ones((500, 1), int)
zero = np.zeros((500, 1), int)

label_ndarray = np.concatenate((zero, one))
np.concatenate((joins, label_ndarray), 1).shape
```
