```python
# 問題5

尖度とは、分布が正規分布からどれだけ尖っているかを表す統計量。 山の尖り度と裾野の広がり度を示す。
歪度とは、分布が正規分布からどれだけ歪んでいるかを表す統計量で、左右対称性を示す指標。
```

```python
# 問題6

import seaborn as sns

sns.distplot(train.iloc[:, -1])
plt.show()
```

```python
"""
seabornのsns.displot()やsns.histplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
"""
print("尖度 :", train.iloc[:, -1].kurt())
print("歪度 :", train.iloc[:, -1].skew())
```

```python
"""
目的変数に対して対数変換を行う。
"""
import numpy as np

target_e = np.log(train.iloc[:, -1])
```

```python
"""
対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。
"""
sns.distplot(target_e)
plt.show()
```

```python
# 問題7

"""
相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。
"""
train_corr = train.corr(method='pearson')
plt.figure(figsize=(20, 15))
sns.heatmap(train_corr, cbar=True, annot=True, square=True,
            fmt='.2f', annot_kws={'size':7})
plt.show()
```

```python
"""
目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。
"""
train_corr_rsort = train_corr.sort_values(train.columns[-1], axis=1, ascending=False)
train_corr_sort = train_corr_rsort.sort_values(train.columns[-1], ascending=False)
plt.figure(figsize=(15,10))
sns.heatmap(train_corr_sort.iloc[1:11, 1:11],
            cbar=True, annot=True, square=True, fmt='.2f',
            annot_kws={'size':20})
plt.show()
```

```python
"""
選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。
"""
OverallQual：全体的な材料と仕上げの品質
GrLivArea：地上リビングエリアの平方フィート
GarageCars：車の容量におけるガレージのサイズ
GarageArea：平方フィート単位のガレージのサイズ
TotalBsmtSF：地下室の総平方フィート
1stFlrSF：1階平方フィート
FullBath：グレード以上のフルバスルーム
TotRmsAbvGrd：グレードを超える客室の合計（バスルームは含まれません）
YearBuilt：元の建設日
YearRemodAdd：リフォーム日
```

```python
"""
選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。
"""
train_corr_10f1di = train_corr_sort.iloc[1:11, 1:11].values.flatten()
corr_top = sorted(set(train_corr_10f1di))[-4:-1]
X = train_corr_sort.iloc[1:11, 1:11][train_corr_sort.
                                  iloc[1:11, 1:11].isin(corr_top)]

indices = np.where(X < 1)
indices_name = [[X.index[x], X.columns[y]] for x, y in zip(*indices)
                if x != y and x < y]
values =  [X.loc[n][m] for n, m in indices_name]
dict_values_indices = dict(zip(values, indices_name))

for a in range(3):
    print('相関係数：', values[a])
    print('特徴量の組み合わせ：', indices_name[a], '\n')
```
