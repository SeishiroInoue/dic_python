```python
# 紙を1回折った時の厚さを計算するコード

THICKNESS = 0.00008

folded_thickness = THICKNESS * 2 ** 43

print("厚さ： {}メートル".format(folded_thickness))
```

```python
# メートルをキロメートルに変換して小数点以下2桁で表示する
print("厚さ： {:.2f}万キロメートル".format(folded_thickness / 10000000))
```

```python
THICKNESS = 0.00008
folded_thickness = []
folded_thickness.append(THICKNESS)
for i in range(43):
  folded_thickness.append(THICKNESS * 2 ** i)
print(len(folded_thickness))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

```python
# 緑色の折れ線グラフを表示する。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness, color='green')
plt.show()

# 太い折れ線グラフを表示する。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness, linewidth=10)
plt.show()

# 点線の折れ線グラフを表示する。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness, linestyle=':')
plt.show()
```
