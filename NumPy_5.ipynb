```python
# 問題 1

import numpy as np

csv_path = "/content/mtfuji_data.csv" # ファイル名（パス）を指定する

np.set_printoptions(suppress=True) # 指数表示の禁止を設定する
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
```

```python
# 問題4

def descend(posi=136, step=0.2):
  s_list = [posi]
  while True:
    posi = posi - step*gradient(posi)
    posi = int(posi.round())
    if 0 <= posi < 300:
      if posi in s_list:
        break
      else:
        s_list.append(posi)
    else:
      break
    return s_list
```

```python
# 問題5

s_point = descend(136, 0.2)

plt.title('descending')
plt.xlabel('position')
plt.ylabel('elevation')
plt.plot(fuji[s_point, 3])

plt.show()
```
